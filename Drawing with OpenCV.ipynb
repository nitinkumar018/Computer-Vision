# We will learn how to use OpenCV to draw:

Lines, 
Rectangles, 
Circles

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
np.zeros(shape=(3,3),dtype=int) # zeros is a function that creates an array having zeroes in it (3 rows, 3 columns, bidefault layer is only one)
# pixel 0 means black image/ dark image
```

```python
# create a blank image using zeros func. which is present in numpy lib. 
#zeros is a function that creates an array having zeroes in it (512 rows, 512 columns, 3 layers i.e. b, g, r)
blank_img=np.zeros(shape=(512,512,3),dtype=np.int16) # bidefault int takes 64 bits but here we pass int16 i.e. we are reducing the size of the image to save the storage
```

```python
blank_img   # pixel 0 means dark image
```

```python
blank_img.shape  # height, width, no. of layers
```

```python
plt.imshow(blank_img)  # visualize the image
```

```python
cv2.imshow("test",blank_img)  # visualize the image on computer vision window test
cv2.waitKey()                 # keep the image on the screen for infinite duration  (we can also mention the time duration)
cv2.destroyAllWindows()       # closes the window automatically (if we mention the time duration)
```

```python
#pt1=top left vertex
#pt2=bottom right vertex
# draw a rectangle on blank image using rectangle func
cv2.rectangle(blank_img,pt1=(384,0),pt2=(510,128),color=(0,255,0),thickness=5)  # color is green
```

```python
plt.imshow(blank_img)  # visualize the image
```

```python
#pt1=top left
#pt2=bottom right
# draw another rectangle 
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)  # color is blue
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(255,0,0),thickness=5) # draw a circle using circle func (color is red)
plt.imshow(blank_img)
```

```python
# draw another circle 
cv2.circle(img=blank_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1) # -ve thinkness means filled circle 
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(0,0),pt2=(511,511),color=(102,255,255),thickness=2) # draw a line using line func
plt.imshow(blank_img)
```
